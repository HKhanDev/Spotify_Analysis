```
import pandas as pd
```

```
data = pd.read_csv('tracks.csv')
```

```
data
```

```
pd.isnull(data).sum()
```

```
data['release_date'] = pd.to_datetime(data['release_date'])
year = data['release_date'].apply(lambda x : x.year)
month = data['release_date'].apply(lambda x : x.month)
data.insert(loc=8, column='year', value=year) 
data.insert(loc=9, column='month', value=month)
```

```
most_popular = data.query('popularity>90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```
pop_date = most_popular.sort_values('release_date', ascending=False)
pop_date[['name', 'popularity', 'explicit','release_date']][:20]
```

```
most_popular_march_20 = data.query('(popularity > 80) and (year in ["2020"]) and (month in ["3"])')
most_popular_march_20
```

```
df_1=data.groupby('popularity')['danceability'].mean().sort_values(ascending=[False]).reset_index()
df_1.head()
```

```
import plotly.express as px   #importing plotly
fig2 = px.scatter(df_1, x="popularity", y="danceability", color="danceability",size='popularity')
fig2.show()
```

```
from scipy.stats import pearsonr    #importing the library
data1 = df_1['popularity']
data2 = df_1['danceability']

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)
print('Pearsons correlation: %.5f' % corr)
```

```
import plotly.graph_objects as go    #importing the library

matrix=data.corr()  #returns a matrix with correlation of all features
x_list=['popularity','duration_ms','explicit',
        'danceability','energy','key','loudness',
        'mode','speechiness','acousticness','instrumentalness',
        'liveness','valence','tempo','time_signature']

fig_heatmap = go.Figure(data=go.Heatmap(
                   z=matrix,
                   x=x_list,
                   y=x_list,
                   hoverongaps = False))
fig_heatmap.update_layout(margin = dict(t=200,r=200,b=200,l=200),
    width = 800, height = 650,
    autosize = False )

fig_heatmap.show()
```
